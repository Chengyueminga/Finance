### 3.3 Normal Returns as the Benchmark Case 
### geometric Brownian motion of BSM given by SDE  
$ds_t = r S_tdt + \sigma S_tdz_t$
### A discrete version, which can easily be simulated, is given by the difference equation  
$S_t = S_{t - \delta t}e^{(r - \frac{1}{2}\sigma ^2)\delta t + \sigma \sqrt{\delta tz_t}}$

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib as mpl
import math
```

```python
###GBM simulation
def simulateGBM(S0,T,r,sigma):
    #S0 = 100 ##initial index level
    #T = 10 ##Time horizon
    #r = 0.05 ##Risk free rate
    #sigma = 0.2 ##Instantaneous volatility
    
    ##simulation parameters
    np.random.seed(250000)
    dateGBM = pd.DatetimeIndex(start = "05-13-2009",
                              end = "05-13-2019",
                              freq = "B")
    M = len(dateGBM) ##time steps
    dt = 1/252
    df = math.exp(-r * dt)
    
    ##stock price path
    ##random numbers
    rand = np.random.standard_normal((M,1))
    ##store St in S(array)
    S = np.zeros_like(rand)
    S[0] = S0
    for i in range(1,M):
        S[i] = S[i - 1] * np.exp((r - 0.5 * sigma ** 2) * dt + sigma * rand[i] * math.sqrt(dt))
        
    gbm = pd.DataFrame(S[:,0],index = dateGBM, columns = ["price"])
    gbm["return"] = np.log(gbm["price"]/gbm["price"].shift(1))
    
    gbm["variance"] = 252 * np.cumsum(gbm["return"] ** 2) / np.arange(len(gbm))
    gbm["volatility"] = np.sqrt(gbm["variance"])
    gbm = gbm.dropna()
    return gbm
```

```python
simulateGBM(100,10,0.05,0.2)
```

```python
plt.plot(simulateGBM(100,10,0.05,0.2).price)
```

```python
### daily price and log returns
def price_return(data):
    #Get the picture size
    plt.figure(figsize = (9,6))
    plt.subplot(211)
    ###plot
    data["price"].plot()
    ###get the title
    plt.ylabel("daily price")
    ###get the grid in background
    plt.grid(True)
    plt.axis("tight")
    
    plt.subplot(212)
    data["return"].plot()
    plt.ylabel("daily return")
    ##3get the grid background
    plt.grid(True)
    plt.axis("tight")
```

```python
data = simulateGBM(100,10,0.05,0.2)
price_return(data)
```

```python
###probability density function of a normal random variable x
def dN(x,mu,sigma):
    z = (x - mu)/sigma
    pdf = np.exp(-0.5 * z ** 2)/math.sqrt(2 * math.pi * sigma ** 2)
    return pdf
```

```python
###histogram of annualized daily log returns
def return_histogram(data):
    ###get the picture size
    ###get the real frequency
    plt.figure(figsize = (9,6))
    x = np.linspace(min(data["return"]),max(data["return"]),100)
    plt.hist(np.array(np.array(data["return"])),bins = 50,normed = True)
    ###get the standard normal distribution
    y = dN(x,np.mean(data["return"]),np.std(data["return"]))
    plt.plot(x,y,linewidth = 2)
    plt.xlabel("log returns")
    plt.ylabel("frequency")
    plt.grid(True)
```

```python
return_histogram(data)
```

```python
###QQ plot of annualized daily log returns
def return_QQplot(data):
    plt.figure(figsize = (9,6))
    sm.qqplot(data["return"],line = "s")
    plt.grid(True)
    plt.xlabel("theoretical line")
    plt.ylabel("sample quantiles")
```

```python
return_QQplot(data)
```

```python
###mean return, volatility(252 days moving = 1 year)
def rolling_statistics(data):
    plt.figure(figsize = (9,6))
    plt.subplot(311)
    meanReturn = data["return"].rolling(252).mean().dropna() * 252
    meanReturn.plot()
    plt.grid(True)
    plt.ylabel("annualized return")
    
    plt.subplot(312)
    sigma = data["return"].rolling(252).std() * math.sqrt(252)
    plt.plot(sigma)
    plt.grid(True)
    plt.ylabel("volatility")
    
```

```python
rolling_statistics(data)
```
