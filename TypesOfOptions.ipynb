### 10.1 Types of options 

### <font color = blue>call option & put option 
    
$call$ $option$
#### gives the holder of the option the right to $buy$ an asset by a certain date for a certain price
$put$ $option$
#### gives the holder the right to $sell$ an asset by a certain date for a certain price
----
* expiration date / maturity date
#### ----the date specified in the contract
* exercise price / strike price
#### ----the price sepcified in the contract
----

### <font color = blue> American options & European options 
$American options$
#### can be exercised at any time up to the expiration date
$European options$
#### can be exercised only on the expiration date itself

### <font color = blue> Call Options
##### Consinder
    An investor buys an European call option
    ---strike price = $100
---shares = 100
#### Suppose
    ---current stock price = $98
---expiration date of the option is in 4 months
---the price of an option to purchase oneshare is $5
    ---the initial investment is $500
    * If the stock price on this date is less than $100, the investor will clearly choose not to exercise.(The investor loses the whole of the initial investment of $500)
    * If the stock price on this date is above $100, the option will be exercised. (If the stock price is $115,the investor can buy 100 shares for $100 per share)


```python
###Figure 10.1 Profit from buying a European call option on one share of a stock. Option price = 5; strike price = 10
from matplotlib import pyplot as plt
import numpy as np
OptionPrice = 5
StrikePrice = 100
S = np.arange(StrikePrice * 0.7, StrikePrice * 1.3,0.5)
##Np.where(logic,a,b) means if logic is True, a; else, b
PayOff = np.where(S > StrikePrice,S - StrikePrice - OptionPrice,0 - OptionPrice)
plt.plot(S,PayOff)
plt.axhline(y = 0, color = "red")
plt.xlabel('Long Call Option')
plt.ylabel('PayOff')
```

```python
def callPayoff(CallOptionPrice,CallStrikePrice):
    S = np.arange(CallStrikePrice * 0.7,CallStrikePrice * 1.3,0.5)
    PayOff = np.where(S > CallStrikePrice,S - CallStrikePrice - CallOptionPrice, 0 - CallOptionPrice)
    plt.plot(S,PayOff)
    plt.axhline(y = 0, color = 'r')
    plt.xlabel('Long Call Option')
    plt.ylabel('PayOff')
```

```python
callPayoff(5,100)
```

### <font color = blue> Put Options
    Whereas the purchaser of a call opion is hoping the stock price will increase, the purchaser of a put option is hoping that iot will decrease.
##### Consinder
    A investor buys a European put option
    ----strike price = $70
----sell 100 shares of a certain stock
##### Suppose
----Current price = $ 65
    ----expiration date of this option is in 3 months  
    ----the price of an option o sell one share is $7
    
*It will be exercised only if the stock price is below $70 on the expiration date    

---------
##### Suppose
----the stock price is $55 on this date

So the investor can buy 100 shares for $55 per share 
and sell the same shares for 70 dollar to realize a gain of 15 dollar per share.

```python
###Figure10.2 Profit from buying a European put option on one share of a stock. Option price = $7; strike price = $70
import numpy as np
from matplotlib import pyplot as plt
putOptionPrice = 7
strikePrice = 70

S = np.arange(strikePrice * 0.7, strikePrice * 1.3, 0.5)
payOff = np.where(strikePrice > S, strikePrice - S - putOptionPrice, 0 - putOptionPrice)
plt.plot(S, payOff)
plt.axhline(y = 0, color = "red")
plt.xlabel("Long Put Option")
plt.ylabel("Payoff")
```

```python
###import numpy as np
###from matplotlib import pyplot as plt
def putPayOff(PutOptionPrice,PutStrikePrice):
    S = np.arange(PutStrikePrice * 0.7, PutStrikePrice * 1.3, 0.5)
    payOff = np.where(S > PutStrikePrice,0 - PutOptionPrice, PutStrikePrice - S - PutOptionPrice)
    plt.plot(S,payOff)
    plt.axhline(y = 0, color = "red")
    plt.xlabel("Long Put Option")
    plt.ylabel("PayOff")
```

```python
putPayOff(7,70)
```
